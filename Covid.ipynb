```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
```

```python
df_covid_brasil = pd.read_csv('brazil_cities_coordinates.csv')
```

# Nivel Mundo

```python
list(df_covid_brasil.columns)
```

```python
df_heat_map = df_covid_brasil[['lat', 'long','city_name']]
```

```python
df_heat_map = df_heat_map.replace(to_replace=r',', value='.', regex=True)
```

```python
df_heat_map.isnull().sum()
```

```python
df_heat_map.columns
```

```python
crime_map = folium.Map(location=[-2.5100598,-44.2511041], 
                       tiles = "Stamen Toner",
                      zoom_start = 4)

data_heatmap = df_heat_map[['lat', 'long', 'city_name']]
data_heatmap = df_heat_map.dropna(axis=0, subset=['lat', 'long'])
data_heatmap = [[row['lat'],row['long']] for index, row in data_heatmap.iterrows()]
HeatMap(data_heatmap, radius=10).add_to(crime_map)

crime_map
```
